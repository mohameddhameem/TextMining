```python
from nltk.corpus import wordnet as wn
```

```python
# To look up a word #lemma of word ,POS , Numbering 
# one menaing of book
wn.synsets('book')
```

```python
# Look up with specified POS - NOUN, VERB, ADJ, ADV
wn.synsets('book', pos = wn.VERB)
```

```python
# Let's examine a synset in more details: its definition, examples, lemma
ss = wn.synsets('book')[0]
```

```python
print(ss.definition())
print(ss.examples())
```

```python
# lemmas - syncsets can be represented by multiple words.
ss.lemmas()
```

```python
lem = ss.lemmas()[0]
```

```python
lem.name()
```

```python
# A synset has related synsets, for example, its hypernyms, hyponyms
ss = wn.synsets('vehicle')[0]
```

```python
ss.hypernyms() # get parent terms
```

```python
ss.hyponyms() # to go down - more specific
```

```python
hyps = list(set([w for s in ss.closure(lambda s:s.hyponyms()) for w in s.lemma_names()]))
hyps
```
