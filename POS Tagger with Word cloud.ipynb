```python
import nltk
from nltk import word_tokenize, pos_tag
```

```python
# ===== POS Tagging using NLTK =====

sent = '''Professor Tan Eng Chye, NUS Deputy President and Provost, and Professor 
Menahem Ben-Sasson, President of HUJ signed the joint degree agreement at NUS, 
in the presence of Ambassador of Israel to Singapore Her Excellency Amira Arnon 
and about 30 invited guests, on July 03, 2013.
'''
```

```python
# The input for POS tagger needs to be tokenized first.
sent_pos = pos_tag(word_tokenize(sent))
sent_pos
```

```python
# A more simplified tagset - universal
sent_pos2 = pos_tag(word_tokenize(sent), tagset='universal')
sent_pos2
```

```python
# The wordnet lemmatizer works properly with the pos given
wnl = nltk.WordNetLemmatizer()
wnl.lemmatize('signed', pos = 'v') # can get first letters as parameter for pos tag
```

```python
import wordcloud
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# 1. Simple cloud
# Generate a word cloud image
# Take note that this function requires text string as input
str = ""
str_ori = ""
for item in sent_pos2:
    if (item[1]) == "NOUN" or print(item[1]) == "ADJ" or print(item[1]) == "VERB" or print(item[1]) == "ADP":
        if(item[1] == "NOUN"):
            str = str + " " + wnl.lemmatize(item[0], pos = 'n') # noun
            str_ori = str_ori + " " + item[0]
        elif(item[1] == "VERB"):
            str = str + " " + wnl.lemmatize(item[0], pos = 'v') # verb   
            str_ori = str_ori + " " + item[0]
        else:
            str = str + " " + item[0]
            str_ori = str_ori + " " + item[0]
```

```python
str
```

```python
wc = WordCloud(background_color="white").generate(str)
```

```python
# Display the generated image:
# the matplotlib way:

plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#if its without lemmatize below word cloud
wc = WordCloud(background_color="white").generate(str_ori)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
